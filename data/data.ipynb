```python
import pandas as pd

# baca file csv, kalo NAMA FILE BEDA ganti aja jadi sesuai nama filenya, contoh: '<namaFile>.csv'
df = pd.read_csv('data brw.csv',
                 on_bad_lines='skip', delimiter=';')

df = df.dropna(how="all")
# buat nampilin doang

print(df)
# df.info()
```

```python
# ngilangin kolom timestamp sama rename semua kolom biar jelas
df.drop('Timestamp\n', axis=1, inplace=True)
df.rename(columns={'Media Sosial (Line & Instagram)\nContoh:\nLine: https://line.me/ti/p/Q1hqgliRYf\nInstagram: instagram.com/csui.2022': 'MedSos', 'Deskripsi Singkat\n(Fun Fact atau kalimat bebas yang deskripsiin kamu)': 'Deskripsi Singkat', 'Interest\n(Contoh: Web Dev, UI/UX, Marvel, Valorant.)': 'Interest', 'Pesan untuk CSUI 22\n(Pesan/harapan kamu kedepannya untuk CSUI 22)': 'Pesan buat CSUI22'}, inplace=True)

# buat nampilin doang
df.head()
```

```python
list_medsos = [x for x in df['MedSos']]
list_line = []
list_line_akhir = []
list_ig = []
list_ig_akhir = []

# ngebagi line sama ig berdasarkan input yang bagus
for i in list_medsos:
    line = False
    ig = False
    lala = i.split()
    if not line and not ig:
        # ini ada-ada be ngasih 2 id line pake atau
        if 'atau' in lala:
            list_line.append(lala[1])
            list_ig.append(lala[5])
        else:
            for y in lala:
                if 'line.me' in y:
                    list_line.append(y)
                    line = True
                if 'instagram.com' in y:
                    list_ig.append(y)
                    ig = True
            if not line:
                list_line.append(None)
            if not ig:
                list_ig.append(None)

# benerin biar jadi instagram.com/...... aja
for kata in list_ig:
    if kata != None:
        if 'instagram:https://www.' in kata.lower():
            list_ig_akhir.append(kata[22:])
        elif 'instagram:https://' in kata.lower():
            list_ig_akhir.append(kata[18:])
        elif 'instagram:http://' in kata.lower():
            list_ig_akhir.append(kata[17:])
        elif 'instagram:' in kata.lower():
            list_ig_akhir.append(kata[10:])
        elif 'https://www.' in kata.lower():
            list_ig_akhir.append(kata[12:])
        elif 'https://' in kata.lower():
            list_ig_akhir.append(kata[8:])
        elif 'http://' in kata.lower():
            list_ig_akhir.append(kata[7:])
        elif 'www.' in kata.lower():
            list_ig_akhir.append(kata[4:])
        else:
            list_ig_akhir.append(kata)
    else:
        list_ig_akhir.append(None)

# benerin biar jadi line.me/..... aja
for kata in list_line:
    if kata != None:
        if 'line:https://' in kata.lower():
            list_line_akhir.append(kata[13:])
        elif ':https://' in kata.lower():
            list_line_akhir.append(kata[9:])
        elif 'https://' in kata.lower():
            list_line_akhir.append(kata[8:])
        elif 'http://' in kata.lower():
            list_line_akhir.append(kata[7:])
        else:
            list_line_akhir.append(kata)
    else:
            list_line_akhir.append(None)

# masukin line dari input gajelas
index = 0
for i in list_line_akhir:
    if i == None:
        if 'line' in list_medsos[index].lower():
            list_i = list_medsos[index].split()
            if list_i[1] == ':':
                try:
                    id_line = f'line.me/ti/p/{list_i[2]}'
                    list_line_akhir[index] = id_line
                except:
                    continue
            elif len(list_i) == 5:
                id_line = f'line.me/ti/p/{list_i[4]}'
                list_line_akhir[index] = id_line
            elif len(list_i) == 3:
                if "ig" in list_i[2]:
                    id_line = f'line.me/ti/p/{list_i[1]}'
                    list_line_akhir[index] = id_line
                else:
                    id_line = f'line.me/ti/p/{list_i[2]}'
                    list_line_akhir[index] = id_line
            elif len(list_i) == 2:
                id_line = f'line.me/ti/p/{list_i[0][5:]}'
                list_line_akhir[index] = id_line
            elif 'instagram' in list_i[1]:
                id_line = f'line.me/ti/p/{list_i[3]}'
                list_line_akhir[index] = id_line
            elif 'instagram' in list_i[0].lower():
                id_line = f'line.me/ti/p/{list_i[3]}'
                list_line_akhir[index] = id_line
            else:
                id_line = f'line.me/ti/p/{list_i[1]}'
                list_line_akhir[index] = id_line
    index += 1

# masukin ig dari input gajelas
id = 0
for i in list_ig_akhir:
    if i == None:
        if 'instagram' in list_medsos[id].lower():
            list_i = list_medsos[id].split()
            if len(list_i) == 6:
                link_ig = f'instagram.com/{list_i[5]}'
                list_ig_akhir[id] = link_ig
            elif len(list_i) == 5:
                link_ig = f'instagram.com/{list_i[4]}'
                list_ig_akhir[id] = link_ig
            elif len(list_i) == 4:
                if 'instagram' in list_i[0].lower():
                    if '@' in list_i[1]:
                        link_ig = f'instagram.com/{list_i[1][1:]}'
                        list_ig_akhir[id] = link_ig
                    elif '/' in list_i[3]:
                        splitan = list_i[1].split('/')
                        link_ig = f'instagram.com/{splitan[1]}'
                        list_ig_akhir[id] = link_ig
                    else:
                        link_ig = f'instagram.com/{list_i[1]}'
                        list_ig_akhir[id] = link_ig
                elif '/' in list_i[3]:
                    splitan = list_i[3].split('/')
                    link_ig = f'instagram.com/{splitan[1]}'
                    list_ig_akhir[id] = link_ig
                else:
                    link_ig = f'instagram.com/{list_i[3]}'
                    list_ig_akhir[id] = link_ig
            elif len(list_i) == 3:
                link_ig = f'instagram.com/{list_i[2]}'
                list_ig_akhir[id] = link_ig
            elif len(list_i) == 2:
                link_ig = f'instagram.com/{list_i[1][10:]}'
                list_ig_akhir[id] = link_ig
            else:
                if '/' in list_i[0]:
                    splitan = list_i[0].split('/')
                    link_ig = f'instagram.com/{splitan[1]}'
                    list_ig_akhir[id] = link_ig
        elif '@' in list_medsos[id].lower():
            link_ig = f'instagram.com/{list_medsos[id][1:]}'
            list_ig_akhir[id] = link_ig
        elif 'ig' in list_medsos[id].lower():
            splitan = list_medsos[id].split()
            link_ig = f'instagram.com/{splitan[2][3:]}'
            list_ig_akhir[id] = link_ig
    id += 1

# ngilangin , : di akhir kalo ada BAGIAN IG
index = 0
for i in list_ig_akhir:
    if i != None:
        if i.endswith(',') or i.endswith(':'):
            list_ig_akhir[index] = i[:-1]
    index += 1

# ngilangin , . : di akhir kalo ada BAGIAN LINE
index = 0
for i in list_line_akhir:
    if i != None:
        if i.endswith(',') or i.endswith('.') or i.endswith(':'):
            list_line_akhir[index] = i[:-1]
    index += 1



# ini buat ngecek aja
# id = 0
# for i in list_ig_akhir:
#     print(id, i)
#     print(id, list_medsos[id])
#     id += 1

# id = 0
# for i in list_line_akhir:
#     print(id, i)
#     print(id, list_medsos[id])
#     id += 1
```

```python
from __future__ import print_function

import os.path

from google.auth.transport.requests import Request
from google.oauth2.credentials import Credentials
from google_auth_oauthlib.flow import InstalledAppFlow
from googleapiclient.discovery import build
from googleapiclient.errors import HttpError
from apiclient import errors
from apiclient.http import MediaFileUpload

# If modifying these scopes, delete the file token.json.
SCOPES = ['https://www.googleapis.com/auth/drive.metadata.readonly']


def main():
    """Shows basic usage of the Drive v3 API.
    Prints the names and ids of the first 10 files the user has access to.
    """
    creds = None
    # The file token.json stores the user's access and refresh tokens, and is
    # created automatically when the authorization flow completes for the first
    # time.
    if os.path.exists('token.json'):
        creds = Credentials.from_authorized_user_file('token.json', SCOPES)
    # If there are no (valid) credentials available, let the user log in.
    if not creds or not creds.valid:
        if creds and creds.expired and creds.refresh_token:
            creds.refresh(Request())
        else:
            flow = InstalledAppFlow.from_client_secrets_file(
                'credentials.json', SCOPES)
            creds = flow.run_local_server(port=0)
        # Save the credentials for the next run
        with open('token.json', 'w') as token:
            token.write(creds.to_json())

    try:
        service = build('drive', 'v3', credentials=creds)
        tt = df["id foto"]
        list_nama_foto = []
        # Call the Drive v3 API
        for i in tt:
            file = service.files().get(fileId=i).execute()
            list_nama_foto.append(file['name'])
        df["nama foto"] = list_nama_foto
        df["nama foto"]
    except HttpError as error:
        # TODO(developer) - Handle errors from drive API.
        print(f'An error occurred: {error}')


main()
```
